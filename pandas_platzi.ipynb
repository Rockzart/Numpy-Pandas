## Group by ##

Al agrupar por ejemplo por autor, este campo se vuelve el índice

```python
df_books.groupby('Author').count()
```

```python
df_books.groupby('Author').max()
```

Para que el campo por el que se agrupa no sea el índice se escribe reset_index

```python
df_books.groupby('Author').count().reset_index()
```

Esto es equivalente:

```python
df_books.groupby('Author', as_index=False).count()
```

La función agg() permite aplicar varias funciones al DataFrame una vez agrupado según una columna específica.

En este caso se agrupa por 'Author' y se muestra el mínimo y máximo de las demás columnas.

```python
df_books.groupby('Author').agg(['min', 'max'])
```

Agrupar por Author, obtener el mínimo y máximo de la columna 'Reviews' y sumar los valores de la columna 'User Rating'

```python
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'})
```

```python
df_books.groupby(['Author','Year']).count()
```

## Combinación de data frames ##

```python
df1 = pd.DataFrame(
    {
        'A':['A0', 'A1', 'A2','A3'],
        'B':['B0', 'B1', 'B2','B3'],
        'C':['C0', 'C1', 'C2','C3'],
        'D':['D0', 'D1', 'D2','D3']
        }
    )
df1
```

```python
df2 = pd.DataFrame(
    {
        'A':['A4', 'A5', 'A6','A7'],
        'B':['B4', 'B5', 'B6','B7'],
        'C':['C4', 'C5', 'C6','C7'],
        'D':['D4', 'D5', 'D6','D7']
        }
    )
df2
```

**CONCAT**

```python
pd.concat([df1, df2])
```

Como el concat deja los índices de los data frames y en este caso se repiten, hay que arreglarlo

```python
pd.concat([df1,df2], ignore_index= True)
```

La concatenación por default es por filas pero si se quiere concatenar por columnas se especifica el axis=1

```python
pd.concat([df1,df2], axis = 1)
```

**MERGE**

```python
izq = pd.DataFrame(
    {
        'key' : ['k0', 'k1', 'k2','k3'],
        'A' : ['A0', 'A1', 'A2','A3'],
        'B': ['B0', 'B1', 'B2','B3']
        }
    )
izq
```

```python
der = pd.DataFrame(
    {
        'key' : ['k0', 'k1', 'k2','k3'],
        'C' : ['C0', 'C1', 'C2','C3'],
        'D': ['D0', 'D1', 'D2','D3']
        }
    )
der
```

```python
izq.merge(der)
```

```python
izq.merge(der, on='key')
```

Qué pasa si se modifica el valor con el que se hace match? En este caso se cambia key a key2 en el data frame 2

```python
der = pd.DataFrame(
    {
        'key2' : ['k0', 'k1', 'k2','k3'],
        'C' : ['C0', 'C1', 'C2','C3'],
        'D': ['D0', 'D1', 'D2','D3']
        }
    )
der
```

Al momento de hacer el merge hay que especificar el nombre de la llave para unir los data frames

```python
izq.merge(der, left_on = 'key', right_on='key2')
```

## Pivot y Melt ##

**pivot table**

Los valores de Author pasan a formar el índice por fila y los valores de Genre pasan a formar parte de los índices por columna, y el User Rating se mantiene como valor.

```python
df_books.pivot_table(index='Author', columns='Genre', values='User Rating')
```

En este caso tenemos por cada género, la suma a lo largo de los años. Esto es mucho más interesante, ¿verdad? La mejor noticia es que no solo podemos obtener la suma, también podemos obtener la media, la desviación estándar, el conteo, la varianza, etc. Únicamente con cambiar el parámetro aggfunc que traduce función de agrupamiento.

```python
df_books.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='sum')
```

**melt**

El método melt toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas para especificar la antigua columna y el valor que traía.

Por ejemplo, al imprimir las cinco primeras filas del DataFrame con las columnas de Name y Genre se tiene este resultado.

```python
df_books[['Name','Genre']].head(5)
```

Ahora cada resultado de las dos columnas pasa a una fila de este modo a tipo llave:valor.

```python
df_books[['Name','Genre']].head(5).melt()
```

Ahora podemos seleccionar las columnas que no quiero hacer melt usando el parámetro id_vars. Para este caso Year y también la única columna que quiero aplicar el melt, para este caso Genre con la propiedad value_vars.

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```

## Apply ##

Se define una función

```python
def two_times(value):
    return value * 2
```

Se aplica la función a la columna 'User Rating' que es multiplicarla por 2

```python
df_books['User Rating'].apply(two_times)
```

Agragar la columna calculada

```python
df_books['User Rating * 2'] = df_books['User Rating'].apply(two_times)
df_books
```

```python
df_books['User Rating * 3'] = df_books['User Rating'].apply(lambda x: x* 3)
df_books
```

Los fiction se multiplican por 2 pero los non fiction no

```python
df_books['User Rating * 2'] = df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis = 1)
df_books
```
